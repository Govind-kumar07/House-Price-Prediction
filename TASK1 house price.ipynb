```python
# Print available columns to help find relevant features
print("Available columns:", train.columns.tolist())
```

```python
# Select features (Update column names if they differ)
# We'll use square footage (e.g., 'GrLivArea'), bedrooms ('BedroomAbvGr'), bathrooms ('FullBath') 
features = ['GrLivArea', 'BedroomAbvGr', 'FullBath']
target = 'SalePrice'
```

```python
# Drop rows with missing values in selected features (basic handling)
train = train.dropna(subset=features + [target])
test = test.fillna(test[features].median())  # Fill test set missing values with median
```

```python
# Define training data
X_train = train[features]
y_train = train[target]
```

```python
# Define test data
X_test = test[features]
```

```python
# Initialize and train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions
predictions = model.predict(X_test)
```
